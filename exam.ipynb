## Investigating how the mentions of body parts are gendered

```python
# Helper function: Identify whether a noun is a body part
with open("bp_list.txt", 'r') as file:
    body_part_list = [line.strip() for line in file]

def is_body_part(word):
    return word.lower() in body_part_list
```

In order to investigate the mentions of body parts, I first write a helper function identifying the body parts in the text. The list of body parts is retrieved from https://github.com/janester/mad_libs/blob/master/List%20of%20Body%20Parts.txt.


```python
# Load spaCy model
nlp = spacy.load('en_core_web_lg')

def spacy_body_part(nlp, text):
    # Get rid of unwanted linebreaks
    cleaned_text = ' '.join(text.split())
    doc = nlp(cleaned_text)

    bp_name = []
    bp_sent = []
    bp_des = []
    bp_pronoun = []
    for sent in doc.sents:
        for token in sent:
            lem = token.lemma_
            person_name = None
            person_pnoun = None
            if not token.is_punct:
                if token.pos_ == "PRON":
                    person_pnoun = token.text
                if token.pos_ != "NOUN":
                    continue
                if token.ent_type_ == "PERSON":
                    person_name = token.text
                # Check if the noun in the sentence is a body part
                if is_body_part(lem):
                    pronoun = []
                    description = []
                    
                    # Check if there is a gender pronoun as indicator
                    for child in token.children:
                        if child.pos_ in ["ADJ", "VERB"]:
                            description.append(child.text)
                        if child.pos_ == "PRON":
                            pronoun.append(child.text)
                        if child.ent_type_ == "PERSON":
                            pronoun.append(child.text)
                    if pronoun == []:
                        if person_name is not None:
                            pronoun.append(person_name)
                        elif person_pnoun is not None:
                            pronoun.append(person_pnoun)
                    # Ensure only body parts with pronouns indicated is stored
                    if pronoun:
                        bp_name.append(lem)
                        bp_sent.append(sent.text)
                        bp_des.append(description)
                        bp_pronoun.append(pronoun)
    
    bp_df = pd.DataFrame({"Body Part": bp_name, "Description": bp_des, "Person":bp_pronoun,  "Sentence": bp_sent})
    return bp_df
```

```python
# Helper function: Run the spacy model and return mentioned gendered body parts
def run_spacy(dataset):
    dfs_list = []
    for index, row in dataset.iterrows():
        title = row['Title']
        content = row['Content']
        result_df = spacy_body_part(nlp, content)
        result_df['Title'] = title
        dfs_list.append(result_df)
        final_df = pd.concat(dfs_list, ignore_index=True)
    return final_df
```

### Frequency of body parts mentioned by gender

```python
# Helper function: Plotting frequency
def frequency_plot(data1, data2, title):
    plt.figure(figsize=(12, 8))
    # Get unique body part counts
    unique_counts = np.unique(list(data1.index) + list(data2.index))
    
    # Create x-axis positions for bars
    x_positions = np.arange(len(unique_counts))

    # Plot female body part counts
    plt.bar(x_positions - 0.2, [data1.get(part, 0) for part in unique_counts], width=0.5, color='lightcoral', label='Female', alpha=0.7)
    # Plot male body part counts
    plt.bar(x_positions + 0.2, [data2.get(part, 0) for part in unique_counts], width=0.5, color='skyblue', label='Male')

    plt.title(title)
    plt.xlabel('Body Part')
    plt.ylabel('Frequency')
    plt.xticks(x_positions, unique_counts, rotation=45, ha='right')
    plt.legend()
    plt.tight_layout()
    plt.show()
```

### Gender skew of body parts

```python
# Helper function: Calculate the skew scores
def body_part_skew(male, female):
    skew_dic = {}
    common_body_parts = set(male.index) & set(female.index)
    
    for body_part in common_body_parts:
        pct_m = male[body_part] / male.sum()
        pct_f = 0

        # Check if the body part is present in the female dataset
        if body_part in female.index:
            pct_f = female[body_part] / female.sum()
        
        # Calculate the skew score
        if pct_m > pct_f:
            skew = pct_m / pct_f
        elif pct_m < pct_f:
            skew = -pct_f / pct_m
        else:
            skew = 0
        skew_dic[body_part] = skew

    return skew_dic, common_body_parts
        

def plot_skew_score(skew_dic, common_body_parts, date):
    # Convert the dictionary to two lists for plotting
    body_parts, scores = zip(*skew_dic.items())

    # Create a scatter plot with labels
    plt.figure(figsize=(12, 8))
    colors = ['red' if score < 0 else 'blue' for score in scores]
    plt.scatter(scores, np.arange(len(common_body_parts)), color=colors, alpha=0.3)

    # Add labels to each point
    for i, (body_part, score) in enumerate(zip(body_parts, scores)):
        plt.text(score, i, body_part, ha='left', va='center')
    
    # Draw a horizontal line at x=0
    plt.axvline(0, color='black', linestyle='--', linewidth=2)

    # Set labels and title
    plt.xlabel('Skew Score')
    plt.title('Skew Scores of Body Parts in ' + date + ' 1800s')

    # Show plot
    plt.show()
```

```python
ss_early, bp_early = body_part_skew(early_male_counts, early_female_counts)
print(ss_early)
plot_skew_score(ss_early, bp_early, "Early")

ss_late, bp_late = body_part_skew(late_male_counts, late_female_counts)
plot_skew_score(ss_late, bp_late, "Late")
```

These two skewed score graphs offer more direct insights into how body parts are gendered in the early and late 19th century by highlighting frequently mentioned body parts with colors. 

Throughout the century, the most distinctly female-associated body part is the eyelash, followed by the waist and cheek, aligning perfectly with societal perceptions of the female body. It is noteworthy that in the early 19th century, female-skewed body parts predominantly concentrate around the face, reinforcing the societal expectation that a woman needs to embody a certain level of beauty. Transitioning into the late 1800s, more female-skewed body parts emerge, with attention shifting away from the face to parts like the spine, ankle, and wrist. However, these body parts do not symbolize power; instead, they convey a sense of fragility, as they are joints or bone parts that could be easily broken.

Conversely, the "masculine body parts" exhibit a more intriguing pattern. In the early 1800s, the most male-associated body part is the heel, potentially linked to the prevalent footwear worn by men during that era, such as shoes with notable heels referred to as "brogue" or "top boots". Considerate attention to footwear reflects the character's status and personality. Heels are also associated with leg movements, resonating with the gender stereotype of men engaging in physical activities in the public sphere and valuing strength and power. Other frequently mentioned body parts for men in the early 1800s contribute to portrait the same image. However, in the late 1800s, there is a shift in male body parts towards those that are not easily visible: organs, muscles, blood, skeleton, and scalp. This change indicates a evolving focus on internal aspects, possibly reflecting a deeper exploration of male identity beyond superficial appearances during this period.

## Predicting the historial period of the text

Now that we have seen that there is a difference in the mentions of body parts for different genders in the early and late 19th century, we will then investigate classification methods to predcit whether a text is written/published in early or late 1800s with this body part feature. Both token-based and word-embedding-based classifiers will be explored.

```python
# Prepare the data and the labels
data_early['Label'] = "Early"
sample_late['Label'] = "Late"
merged_df = pd.concat([data_early, sample_late], ignore_index=True)
merged_df = shuffle(merged_df)
merged_df
```

```python
# For the model, we need the labels and the text
text_label = merged_df['Label'].tolist()
text_content = merged_df['Content'].tolist()
```

### Token-Based Classifier

```python
# Vectorize the data
vectorizer = TfidfVectorizer()
fiction_fm = vectorizer.fit_transform(text_content)
scaled_fiction = StandardScaler().fit_transform(fiction_fm.toarray())

# Select the 20 most informative features
selector = SelectKBest(mutual_info_classif, k=20)
selected_fm = selector.fit_transform(scaled_fiction, text_label)

# Split the dataset
X_train, X_test, y_train, y_test = train_test_split(selected_fm, text_label, test_size=0.2, random_state=42)

# Initialize classifier
log_clf = LogisticRegression(max_iter=1000)

# Train the classifier
log_clf.fit(X_train, y_train)

# Make predictions on the test set
y_pred = log_clf.predict(X_test)

# Evaluate the classifier on the test set
accuracy = accuracy_score(y_test, y_pred)
print(f"Test Set Accuracy Score: {accuracy:.4f}")

# Calculate a 5-fold cross-validated accuracy score
token_cross_val = cross_val_score(log_clf, selected_fm, text_label, cv=5, scoring='accuracy')

# Print the resulting cross-validated score
mean_accuracy = np.mean(token_cross_val)
print(f"5-fold Cross-Validated Classification Score: {mean_accuracy:.4f}")
```

Now we have ran the baseline model for a token-based logistic classifer with feature matrix built by the `SelectKBest` method. This allows us to select the top k most informative features based on a scoring function. The model has a fairly good predicting score of 0.9655 and a very high cross-validated score of 0.9714. These scores both suggest that this classifer has a great performance accurately capturing the patterns of the data.


```python
# Helper Function: Calculate feature values based on skew score of body parts
def calculate_body_part_features(text_data, skew_scores):
    feature_matrix = np.zeros((len(text_data), len(skew_scores)))
    for i, text in enumerate(text_data):
        for j, (body_part, skew_score) in enumerate(skew_scores.items()):
            if body_part.lower() in text.lower():
                feature_matrix[i, j] = skew_score
    return feature_matrix
```

```python
# Mannually build a feature matrix about gendered body parts
late_fm = calculate_body_part_features(text_content, ss_late)

# Split the data
X_train_bpfm, X_test_bpfm, y_train_bpfm, y_test_bpfm = train_test_split(late_fm, text_label, test_size=0.2, random_state=42)

# Train the classifier
log_clf.fit(X_train_bpfm, y_train_bpfm)

# Make predictions on the test set
y_pred_bpfm = log_clf.predict(X_test_bpfm)

# Evaluate the classifier on the test set
accuracy = accuracy_score(y_test_bpfm, y_pred_bpfm)
print(f"Test Set Accuracy Score: {accuracy:.4f}")

token_late_bp_cross_val = cross_val_score(log_clf, late_fm, text_label, cv=5, scoring='accuracy')
# Print the resulting cross-validated score
mean_accuracy = np.mean(token_late_bp_cross_val)
print(f"5-fold Cross-Validated Classification Score Using Body Parts as Features: {mean_accuracy:.4f}")
```

From the information above, it is evident that `SelectKBest` outperforms the custom feature matrix based on body parts. The classifier using selected body parts as features yields an accuracy score of 0.6207 and a cross-validated score of 0.6244. ### Word-Embedding-Based Classifier

```python
# Write function returning a word embedding vector

def get_doc_embedding(text, nlp):
    # Process the input text using spaCy
    doc = nlp(text)

    # Initialize variables to store the sum of word vectors and the count of non-stopword tokens
    sum_vectors = np.zeros(doc.vector.shape)
    count = 0

    # Iterate through the tokens
    for token in doc:
        # Remove any token that spaCy tags that is stop word, punct, space or does not have a vector
        if not token.is_stop and not token.is_punct and not token.is_space and token.has_vector:
            # Add the word vector to the sum
            sum_vectors += token.vector
            count += 1

    # Calculate the average word vector by dividing the sum by the count
    if count > 0:
        avg_vector = sum_vectors / count
        return avg_vector
    else:
        return None
```

```python
# Pass each text into the function
embeddings = []
for volumes in text_content:
    embedding = get_doc_embedding(volumes, nlp)
    embeddings.append(embedding)
    
# Store as a feature matrix
embed_fm = np.array(embeddings)

# Print the shape
print("Shape of the feature matrix:", embed_fm.shape)
```

```python
# Split the dataset
X_train_embed, X_test_embed, y_train_embed, y_test_embed = train_test_split(embed_fm, text_label, test_size=0.2, random_state=42)

# Train the classifier
log_clf.fit(X_train_embed, y_train_embed)

# Make predictions on the test set
y_pred_embed = log_clf.predict(X_test_embed)

# Evaluate the classifier on the test set
accuracy_embed = accuracy_score(y_test_embed, y_pred_embed)
print(f"Test Set Accuracy Score: {accuracy_embed:.4f}")
```

```python
# Compute a 5-fold cross-validated logistic regression classification score
embed_cross_val = cross_val_score(log_clf, embed_fm, text_label, cv=5, scoring='accuracy')

print(f"5-fold Cross-Validated Classification Score: {embed_cross_val.mean():.4f}")
```

The base model for word-embedding-based classifer has a predicting accuracy score of 0.8276 and a 5-fold cross-validated score of 0.8724. These scores are not bad, but they definitely needs fine-tuning compared to the scores of the token-based classifer. And next, I will use the body part skew scores as references to build the word embeddings to enhance the model.

```python
# Create a word embedding vector with body part skew score
def get_doc_embedding_with_body_parts(text, nlp, skew_dic):
    # Process the input text using spaCy
    doc = nlp(text)
    word_embeddings = []

    for token in doc:
        word = token.text.lower()

        # Check if the word is in the  body parts skew score dictionary
        if word in skew_dic:
            # Use the  skew score for this body part
            skew_score = skew_dic[word]
            embedding = np.array([skew_score] * 300) # times the dimension
        else:
            # Use the default word embedding
            embedding = token.vector

        word_embeddings.append(embedding)

    # Calculate the embedding
    doc_embedding = np.mean(word_embeddings, axis=0)
    
    return doc_embedding

```

The mechanisms behind these two types of classification help to explain the different outcomes after tuning. Token-based classification operates on individual tokens, and the features are often derived from the presence or absence of these tokens. This results in high-dimensional, sparse feature vectors. However, for word embedding classification, it observes the relationships between words in a continuous vector space, allowing for more nuanced, lower-dimensional, and dense vectors.

To elaborate, the word-embedding-based classifier may comprehend the underlying connections between "waist" and "eyelashes" as they both point towards a feminine beauty standard, whereas for the token-based classifier, these two body parts are sparse features, and the presence or absence of one is not related to the others. As a result, the investigation indicates that the word-embedding-based classifier modified by body part skew scores is more effective in determining the approximate historical period of the fictions, as it understands the semantic relationships between the body parts.
